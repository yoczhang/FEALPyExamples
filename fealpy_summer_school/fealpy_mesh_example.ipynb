# Fealpy 中的网格

```python
import numpy as np 
import matplotlib.pyplot as plt 
from mpl_toolkits.mplot3d import Axes3D
%matplotlib inline
```

## 网格对象

```python
from fealpy.mesh import TriangleMesh

node = np.array([(0.0, 0.0), (1.0, 0.0), (1.0, 1.0), (0.0, 1.0)], dtype=np.float64)
cell = np.array([(1, 2, 0), (3, 0, 2)], dtype=np.int_)
mesh = TriangleMesh(node, cell)

node = mesh.entity('node')
edge = mesh.entity('edge')
cell = mesh.entity('cell')

edge2cell = mesh.ds.edge_to_cell()

print('edge2cell:\n')
for i, val in enumerate(edge2cell):
    print(i, ": ", val)
    
print('edge:\n')
for i, val in enumerate(edge):
    print(i, ": ", val)
    
print('edge:\n')
for i, val in enumerate(cell):
    print(i, ": ", val)

if False:
    print('node:\n', node)
    print('edge:\n', edge)
    print('cell:\n', cell)

fig = plt.figure()
axes = fig.gca()
mesh.add_plot(axes)
mesh.find_node(axes, showindex=True)
mesh.find_edge(axes, showindex=True)
mesh.find_cell(axes, showindex=True)
```

## `MeshFactory`  模块

```python
from fealpy.mesh import MeshFactory

mf = MeshFactory()
```

```python
box = [0, 1, 0, 1]
mesh = mf.boxmesh2d(box, nx=4, ny=4, meshtype='tri')
fig = plt.figure()
axes = fig.gca() 
mesh.add_plot(axes)
```
